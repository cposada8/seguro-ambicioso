```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
# funciones auxiliares
def lanzar_dado(n=6):
    resultado = random.randint(1,n)
    return resultado

def desea_lanzar(respuesta):
    # función para procesar la respuesta en texto de un usuario
    # si retorna algo positivo, la función retorna True
    # si el usuario dice algo negativo, la función retorna False
    
    respuesta = respuesta.lower()
    if respuesta in ["si", "sí", "yes", "true", "y", "s"]:
        return True
    else:
        return False
```

```python
class Agent:
    def __init__(self, name, max_tiros, max_score):
        self.name = name
        self.max_tiros = max_tiros
        self.max_score = max_score
        
    def __str__(self):
        cadena= f"""
        nombre: {self.name},
        max_tiros: {self.max_tiros},
        max_score: {self.max_score}
        """
        return cadena
        
    def desea_lanzar(self, lanzamientos, puntos_turno):
        if lanzamientos <= self.max_tiros and puntos_turno <= self.max_score:
            return True
        else:
            return False

a1 = Agent("a1", 60, 10)

print(a1.name)
print(a1)
a2 = Agent("a2", 5, 100)
```

```python
def jugar_turno(agente, left_score):
    lanzar = True
    perdio = False
    lanzamientos = 0
    puntos_turno = 0
    resultados = []
    while(lanzar and not perdio):
        lanzamientos +=1
        
        resultado = lanzar_dado()
        resultados.append(resultado)
        puntos_turno += resultado

        if resultado == 1:
            puntos_turno = 0
            perdio = True
            
        #print(lanzamientos, resultado, puntos_turno )
        if puntos_turno>= left_score: # parar de lanzar pues ya se alcanzó el objetivo
            break
            
        if (not perdio):
            lanzar = agente.desea_lanzar(lanzamientos, puntos_turno)#desea_lanzar(input("¿desea lanzar de nuevo?"))
            
    return puntos_turno, lanzamientos, resultados
    
```

```python
def jugar_partida(numero_partida, agente, target_score=100):
    turnos = 0
    acumulado = 0

    resultados_turnos = {}
    df_turnos = pd.DataFrame()

    while (acumulado < target_score):
        turnos+=1
        puntos_turno, lanzamientos, resultados = jugar_turno(agente, target_score-acumulado)
        acumulado += puntos_turno

        dict_turno = {
            'partida': numero_partida,
            'agente': agente,
            'agent_name': agente.name,
            'target': target_score,
            'turno': turnos,
            'lanzamientos': lanzamientos,
            'puntos_turno': puntos_turno,
            'resultados': [resultados],
            'acumulado': acumulado
        }
        df_turno = pd.DataFrame(dict_turno)
        df_turnos = df_turnos.append(df_turno)
    return df_turnos, turnos, acumulado
```

```python
# Simular n partidas
def simular_n_partidas(agente, num_sim=100, verbose = 0):
    df_detalle = pd.DataFrame()
    for i in range(num_sim):
        if verbose!=0:
            if i%verbose==0:
                print("simulando caso ", i)

                
        df_resultados, turnos, acumulado = jugar_partida(i, agente)
        df_detalle = df_detalle.append(df_resultados)
    
    df_grouped = df_detalle.groupby("partida").agg({'turno':  'max', 
                                                  'lanzamientos': 'mean', 
                                                  'puntos_turno': 'mean',
                                                  'acumulado':'max'})
    return df_grouped, df_detalle
```
